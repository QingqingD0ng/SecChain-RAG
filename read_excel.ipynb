```python
from typing import Any
import pandas as pd
from pydantic import BaseModel
```

```python
# Specify the path to the input XLSX file
input_file = 'Data/Guidelines/Ergebnis_des_IT-Grundschutz-Checks.xlsx'

# Read the data from the XLSX file
df = pd.read_excel(input_file)

class Element(BaseModel):
    type: str
    text: Any
    source: str

    def to_dict(self):
        return {
            "type": self.type,
            "text": self.text,
            "source": self.source 
        }
```

```python
elements = []
for index, row in df.iterrows():
    # Convert the row to a string with columns separated by newlines
    row_string = '\n'.join(f"{col}: {row[col]}" for col in df.columns)
    elements.append([Element(type='xlsx', text=row_string, source='excel')])
```
